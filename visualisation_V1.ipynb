```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exploration des données

```python
Data = pd.read_csv(r"Skyserver.csv")
```

```python
Data.head(10)
```

```python
Data.describe()
```

# Distribution des données dans les différentes classes

```python
#on récupère la colonne classe dans une variale
classe=Data["class"]
```

```python
plt.figure(figsize=(8,8))
plt.pie(classe.value_counts(),labels=classe.value_counts().index,autopct=lambda p:f'{p:.2f}%',
        shadow=True,colors=['mediumvioletred','gold','teal'],labeldistance = 1.1,textprops={'fontsize': 14})

plt.title("Distribution des données dans les différentes classes",fontsize = 15, color="black")

plt.savefig("distribution des données dans les différentes classes.png")
plt.show()
```

# Visualisation de la fréquence d'apparition (d'une valeur) d'attribut (par classe)

```python
data_galaxy=Data[Data['class']=='GALAXY']
data_star=Data[Data['class']=='STAR']
data_quasar=Data[Data['class']=='QSO']
```

```python
features = list(Data.columns[1:])

features.remove('class')
features
```

```python
#pour bien voir , il faut regarder les images qui sont sauvegardées
for i in features:
    plt.figure(figsize=(10,12))
    plt.subplot(3,1,1)
    data_galaxy[i].hist(bins = 10, color="mediumvioletred")
    plt.title("GALAXY",fontsize = 15, color="mediumvioletred")
    plt.subplot(3,1,2)
    data_star[i].hist(bins = 10, color="gold")
    plt.title("STAR",fontsize = 15, color="gold")
    plt.subplot(3,1,3)
    data_quasar[i].hist(bins = 10,color = "teal")
    plt.title("QUASAR",fontsize = 15, color="teal")
    plt.suptitle(i, fontsize = 20, x = 0.5, y = 0.94)
    plt.savefig(i)
```

# Deuxième version avec la fonction de Lili

```python
for i in features:
    plt.figure(figsize=(12,6))
    sns.histplot(data = Data,x = Data[i], hue = 'class', palette='rocket')
   
```
